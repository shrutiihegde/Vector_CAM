```python
import pandas as pd
import numpy as np
```

# Master Sheet 1

```python
m1_df = pd.read_excel('M1.xlsx',  header=[1]) 
m1_df = pd.DataFrame(m1_df)
```

```python
m1_df
```

```python
m1_df[['MOSQUITO ID (Team)', 'Morph ID Species', 'Picture?', 'Abdomen Status', 'Sex']]
```

```python
for col in m1_df.columns:
    print(col)
```

```python
req_cols = ['MOSQUITO ID (Team)', 'Morph ID Species', 'Picture?', 'Abdomen Status', 'Sex']
```

```python
m1_df.loc[:, req_cols]
```

```python
not_null_m1_df = m1_df.loc[(~m1_df['MOSQUITO ID (Team)'].isnull()) & (~m1_df['Morph ID Species'].isnull()) & (~m1_df['Picture?'].isnull()) & (~m1_df['Abdomen Status'].isnull()) & (~m1_df['Sex'].isnull())]
```

```python
m1_df_filtered = not_null_m1_df.loc[:, req_cols]
m1_df_filtered
```

```python
result_m1_df = m1_df_filtered[m1_df_filtered['Picture?'] == 'Y']
```

```python
result_m1_df
```

```python
mosquito_ids = result_m1_df['MOSQUITO ID (Team)'].tolist()
print("Number of Mosquito's: ", len(mosquito_ids))
mosquito_ids
```

```python
print("Morph ID Species: ", result_m1_df['Morph ID Species'].value_counts().to_dict())
print("Abdomen Status: ", result_m1_df['Abdomen Status'].value_counts().to_dict())
print("Number of Males and Females: ", result_m1_df['Sex'].value_counts().to_dict())
```
